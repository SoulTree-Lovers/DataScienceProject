### 모든 OTT 플랫폼 데이터셋을 병합하여 분석해보자 !

```python
# 라이브러리
import pandas as pd
import numpy as np
```

```python
# 파일 불러오기

df_netflix = pd.read_csv("../Netflix_dataset/titles.csv")
df_hbo = pd.read_csv("../HBO_dataset/titles.csv")
df_apple = pd.read_csv("../AppleTV_dataset/titles.csv")
df_disney = pd.read_csv("../Disney_dataset/titles.csv")
df_paramount = pd.read_csv("../Paramount_dataset/titles.csv")
df_amazon = pd.read_csv("../Amazon_dataset/titles.csv")

# pd.set_option('display.max_rows', None) ## 모든 행 출력
```

```python
# 빈 id 행 찾기
print(df_netflix['id'].isnull().value_counts())
print(df_hbo['id'].isnull().value_counts())
print(df_apple['id'].isnull().value_counts())
print(df_disney['id'].isnull().value_counts())
print(df_paramount['id'].isnull().value_counts())
print(df_amazon['id'].isnull().value_counts())
```

-> 비어있는 id는 없음

```python
# 병합 전 데이터 개수 확인
print(len(df_netflix))
print(len(df_hbo))
print(len(df_apple))
print(len(df_disney))
print(len(df_paramount))
print(len(df_amazon))

print(len(df_netflix) + len(df_hbo) + len(df_apple) + len(df_disney) + len(df_paramount) + len(df_amazon))
```

```python
# 각 데이터프레임에 속하는 열을 1로 표시하는 새로운 열 추가
df_netflix['Netflix'] = 1
df_hbo['HBO'] = 1
df_disney['Disney'] = 1
df_apple['Apple TV'] = 1
df_paramount['Paramount'] = 1
df_amazon['Amazon Prime'] = 1
```

```python
# 모든 OTT 데이터프레임 병합 (중복 id 처리 x)
from functools import reduce
pd.set_option('display.max_rows', 10) ## 모든 행 출력

dfs = [df_netflix, df_hbo, df_apple, df_disney, df_paramount, df_amazon]

merged_df = reduce(lambda left, right: pd.merge(left, right, how='outer', on='id'), dfs)
merged_df
```
