```python
# ìˆ˜ì¹˜í˜• ì—´ì˜ ë¶„í¬ ì‹œê°í™”
numeric_columns = df.select_dtypes(include='number').columns
for column in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[column].dropna())
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
# ë²”ì£¼í˜• ì—´ì˜ ë¹ˆë„ìˆ˜ ì‹œê°í™” (id, title, description ì—´ ì œì™¸)
categorical_columns = df.select_dtypes(include='object').columns
categorical_columns = [column for column in categorical_columns if column not in ['id', 'title', 'description', 'genres']]
```

```python
for column in categorical_columns:
    plt.figure(figsize=(10, 4))
    sns.countplot(data=df, x=column, order=df[column].value_counts().index)
    plt.title(f"Frequency of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.xticks(rotation=45)
    plt.show()
```

```python
# genre ì—´ì˜ ë°ì´í„°ë¥¼ ì‰¼í‘œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ì—¬ í•©ì§‘í•© ë§Œë“¤ê¸°
genres = set()
for genre_list in df['genres']:
    if pd.notnull(genre_list):
        genres.update(genre_list.split(','))
```

```python
# ê° ë¦¬ìŠ¤íŠ¸ì˜ ì›ì†Œë¥¼ í¬í•¨í•˜ëŠ” ë¹ˆë„ìˆ˜ ê³„ì‚°
genre_counts = {genre: 0 for genre in genres}
for genre_list in df['genres']:
    if pd.notnull(genre_list):
        for genre in genre_list.split(','):
            genre = genre.strip()
            if genre in genre_counts:
                genre_counts[genre] += 1
            else:
                genre_counts[genre] = 1
```

```python
# ë¹ˆë„ìˆ˜ ì‹œê°í™”
plt.figure(figsize=(12, 6))
sns.barplot(x=list(genre_counts.values()), y=list(genre_counts.keys()))
plt.title("Frequency of Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
import re

# genre ì—´ì˜ ë°ì´í„°ë¥¼ ì‰¼í‘œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ì—¬ ì•ŒíŒŒë²³ë§Œ ë‚¨ê¸´ ë¬¸ìžì—´ë¡œ ë³€í™˜í•˜ì—¬ í•©ì§‘í•© ë§Œë“¤ê¸°
genres = set()
for genre_list in df['genres']:
    if pd.notnull(genre_list):
        genres.update(re.findall(r'[a-zA-Z]+', genre_list))


# ë³´ê¸° ì‰½ê²Œ ì•ŒíŒŒë²³ ìˆœ ì •ë ¬
genres = sorted(list(genres))

        
# ê° ë¬¸ìžì—´ì˜ ë¹ˆë„ìˆ˜ ê³„ì‚°
genre_counts = {genre: 0 for genre in genres}
for genre_list in df['genres']:
    if pd.notnull(genre_list):
        for genre in re.findall(r'[a-zA-Z]+', genre_list):
            genre = genre.lower()
            genre_counts[genre] += 1

# ë¹ˆë„ìˆ˜ ì‹œê°í™”
plt.figure(figsize=(12, 6))
sns.barplot(x=list(genre_counts.values()), y=list(genre_counts.keys()))
plt.title("Frequency of Genres")
plt.xlabel("Count")
plt.ylabel("Genre")
plt.show()
```

```python
# genre ì—´ì˜ ë°ì´í„°ë¥¼ ì‰¼í‘œë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë¶„ë¦¬í•˜ì—¬ ì•ŒíŒŒë²³ë§Œ ë‚¨ê¸´ ë¬¸ìžì—´ë¡œ ë³€í™˜í•˜ì—¬ í•©ì§‘í•© ë§Œë“¤ê¸°
genres = set()
for genre_list in df['production_countries']:
    if pd.notnull(genre_list):
        genres.update(re.findall(r'[a-zA-Z]+', genre_list))
```

```python
# ê° ë¬¸ìžì—´ì˜ ë¹ˆë„ìˆ˜ ê³„ì‚°
genre_counts = {genre: 0 for genre in genres}
for genre_list in df['production_countries']:
    if pd.notnull(genre_list):
        for genre in re.findall(r'[a-zA-Z]+', genre_list):
            genre = genre.lower()
            genre_counts[genre] += 1
```

```python
# ë¹ˆë„ìˆ˜ ì‹œê°í™”
plt.figure(figsize=(12, 6))
sns.barplot(x=list(genre_counts.values()), y=list(genre_counts.keys()))
plt.title("Frequency of Genres")
plt.xlabel("Count")
plt.ylabel("Country")
plt.show()
```
